```python
import pandas as pd
import random
from collections import defaultdict

# Extended list of real restaurant names for added variety
real_restaurant_names = [
    "McDonald's", "Subway", "Starbucks", "Pizza Hut", "Burger King",
    "Domino's", "Dunkin'", "Taco Bell", "KFC", "Chipotle", "Wendy's",
    "Chick-fil-A", "Panera Bread", "Popeyes", "Arby's", "Sonic Drive-In",
    "Little Caesars", "Tim Hortons", "Shake Shack", "Five Guys",
    "Olive Garden", "Red Lobster", "Cheesecake Factory", "IHOP",
    "Applebee's", "Texas Roadhouse", "Outback Steakhouse", "Buffalo Wild Wings",
    "Cracker Barrel", "Chili's", "TGI Friday's", "Ruby Tuesday"
]

# Load the addr_location data from the provided CSV file
from google.colab import files
uploaded = files.upload()
addr_location_data = pd.read_csv('addr_location.csv')

# Extract all unique addresses for foreign key references
addresses = addr_location_data['address'].tolist()

# Determine addresses to be reused twice (50 randomly selected addresses)
reuse_addresses = random.sample(addresses, 50)
address_counts = defaultdict(int)  # Track how many times each address is used

# Franchise configuration
max_franchises = int(300 * 0.40)  # 40% of all tuples can be franchises
franchise_count = 0
restaurant_name_counts = defaultdict(int)  # Track how many times each name is used
restaurant_data = []

# Generate 300 restaurant tuples
while len(restaurant_data) < 300:
    # Assign an address: either one to be reused or a unique one
    address = None
    while True:
        address = random.choice(addresses)
        if address_counts[address] < 2:  # Ensure no address is used more than twice
            address_counts[address] += 1
            break

    # Determine if this is a franchise or a unique restaurant
    if franchise_count < max_franchises and random.random() < 0.40:
        # Select a franchise name, ensuring it's not overused
        potential_names = [
            name for name in real_restaurant_names if restaurant_name_counts[name] < 10
        ]
        if potential_names:
            restaurant_name = random.choice(potential_names)
            franchise_count += 1
        else:
            restaurant_name = random.choice(real_restaurant_names)
    else:
        # Assign a unique restaurant name
        while True:
            restaurant_name = random.choice(real_restaurant_names)
            if restaurant_name_counts[restaurant_name] < 10:
                break

    # Ensure the combination of `restaurant_Name + address` is unique
    if any(
        entry["restaurant_Name"] == restaurant_name and entry["address"] == address
        for entry in restaurant_data
    ):
        continue  # Skip if the combination already exists

    # Assign avgRating, allowing up to 10% NULL values
    avg_rating = round(random.uniform(1.0, 5.0), 2) if random.random() > 0.10 else None

    # Add the restaurant to the dataset
    restaurant_data.append({
        "restaurant_Name": restaurant_name,
        "address": address,
        "avgRating": avg_rating
    })
    restaurant_name_counts[restaurant_name] += 1

# Convert to DataFrame
df_restaurant = pd.DataFrame(restaurant_data)

# Ensure blank entries for `NULL` in the CSV output
df_restaurant['avgRating'] = df_restaurant['avgRating'].apply(lambda x: '' if pd.isnull(x) else x)

# Display the generated data
print(df_restaurant)

# Export to CSV
csv_file_path = "restaurant.csv"
df_restaurant.to_csv(csv_file_path, index=False)
print(f"\nNew data exported to {csv_file_path}")
```

```python
import pandas as pd

# Load the datasets
from google.colab import files
uploaded = files.upload()
population_data = pd.read_csv('population_total.csv')
gdp_data = pd.read_csv('GDP_per_capita.csv')
migration_data = pd.read_csv('net_migration_rate.csv')

# Inspect column names
print("Population Columns:", population_data.columns)
print("GDP Columns:", gdp_data.columns)
print("Migration Columns:", migration_data.columns)

# Rename columns to consistent names for merging
population_data.rename(columns={'slug': 'country_name', 'value': 'population'}, inplace=True)
gdp_data.rename(columns={'slug': 'country_name', 'value': 'gdp_per_capita'}, inplace=True)
migration_data.rename(columns={'slug': 'country_name', ' migrant': 'migration_rate'}, inplace=True)

# Clean the datasets to ensure consistency
# Remove commas and format population as integers
population_data['population'] = population_data['population'].str.replace(',', '').astype(int)

# Remove dollar signs and commas, and format GDP per capita as floats
gdp_data['gdp_per_capita'] = gdp_data['gdp_per_capita'].str.replace('[\$,]', '', regex=True).astype(float)

# Migration rate is already in a numeric format; no changes required

# Merge the datasets
combined_data = population_data.merge(gdp_data, on='country_name', how='inner')
combined_data = combined_data.merge(migration_data[['country_name', 'migration_rate']], on='country_name', how='inner')

# Calculate GDP from GDP per capita and population
combined_data['gdp'] = (combined_data['gdp_per_capita'] * combined_data['population']).astype(int)

# Select and rename the required columns for the `country` table
country_data = combined_data[['country_name', 'population', 'migration_rate', 'gdp']].copy()
country_data.rename(columns={'migration_rate': 'emmigration_rate'}, inplace=True)

# Ensure no negative GDP values (constraint in table schema)
country_data = country_data[country_data['gdp'] > 0]

# Display the resulting dataset
print("\nCombined Country Dataset:")
print(country_data.head())

# Export to CSV
csv_file_path = "country.csv"
country_data.to_csv(csv_file_path, index=False)
print(f"\nNew data exported to {csv_file_path}")
```

demographic_stat

```python
import pandas as pd
import random
from datetime import datetime, timedelta
from google.colab import files

# Upload the region.csv file
uploaded = files.upload()

# Load the region dataset
region_data = pd.read_csv('region.csv')

# Extract region names and population sizes
regions = region_data[['region_Name', 'populationSize']].to_dict('records')

# Generate unique demographic_stat data
num_records = 3000  # Define the number of tuples to generate
demographic_stat_data = []

# Ensure unique ids and (region + datecollected) combinations
used_ids = set()
used_region_date = set()

# Calculate percentages for NULL constraints
max_null_income = int(num_records * 0.0005)  # 0.05% of tuples
max_null_agegroup = int(num_records * 0.0005)  # 0.05% per age group
max_null_multiple_agegroups = int(num_records * 0.0001)  # 0.01% of tuples

null_income_count = 0
null_agegroup_counts = {'youth': 0, 'youngadult': 0, 'middleaged': 0, 'senior': 0}
null_multiple_agegroups_count = 0

start_date = datetime(2020, 1, 1)  # Start of data collection period

while len(demographic_stat_data) < num_records:
    # Generate a unique id
    id_value = random.randint(1, 1000000)
    if id_value in used_ids:
        continue

    # Select a random region
    region = random.choice(regions)
    region_name = region['region_Name']
    population_size = region['populationSize']

    # Generate a random date
    datecollected = start_date + timedelta(days=random.randint(0, 1000))

    # Ensure (region + datecollected) is unique
    if (region_name, datecollected) in used_region_date:
        continue

    # Determine null values for age groups and income level
    age_groups = ['youth', 'youngadult', 'middleaged', 'senior']
    null_groups = random.sample(age_groups, random.choice([0, 1, 2]) if null_multiple_agegroups_count < max_null_multiple_agegroups else 0)
    if len(null_groups) > 1:
        null_multiple_agegroups_count += 1

    for group in null_groups:
        if null_agegroup_counts[group] >= max_null_agegroup:
            null_groups.remove(group)

    # Generate age group percentages
    if 'youth' not in null_groups:
        youth = round(random.uniform(0.0, 0.4), 2)
    else:
        youth = None

    remaining_percentage = 1.0 if youth is None else (1.0 - youth)
    if 'youngadult' not in null_groups:
        youngadult = round(random.uniform(0.0, min(0.4, remaining_percentage)), 2)
    else:
        youngadult = None

    remaining_percentage -= youngadult if youngadult is not None else 0
    if 'middleaged' not in null_groups:
        middleaged = round(random.uniform(0.0, min(0.6, remaining_percentage)), 2)
    else:
        middleaged = None

    senior = None if 'senior' in null_groups else round(1.0 - (youth or 0) - (youngadult or 0) - (middleaged or 0), 2)

    # Generate income level
    income_level = None if null_income_count < max_null_income and random.random() < 0.0005 else random.randint(30000, 120000)
    if income_level is None:
        null_income_count += 1

    # Add the tuple
    demographic_stat_data.append({
        "id": id_value,
        "region": region_name,
        "datecollected": datecollected.date(),
        "youth": youth,
        "youngadult": youngadult,
        "middleaged": middleaged,
        "senior": senior,
        "incomeLevel": income_level
    })
    used_ids.add(id_value)
    used_region_date.add((region_name, datecollected))

# Convert to DataFrame
df_demographic_stat = pd.DataFrame(demographic_stat_data)

# Ensure blank entries for `NULL` in the CSV output
for column in ['youth', 'youngadult', 'middleaged', 'senior', 'incomeLevel']:
    df_demographic_stat[column] = df_demographic_stat[column].apply(lambda x: '' if pd.isnull(x) else x)

# Display the generated data
print(df_demographic_stat.head())

# Export to CSV
csv_file_path = "demographic_stat.csv"
df_demographic_stat.to_csv(csv_file_path, index=False)
print(f"\nNew data exported to {csv_file_path}")
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta
from google.colab import files

# Upload the users.csv file
uploaded = files.upload()

# Load the users dataset
users_data = pd.read_csv('users.csv')

# Extract user IDs
user_ids = users_data['userid'].tolist()

# Define query functions for variety
query_functions = [
    "Analyze Traffic Patterns", "Generate Population Report",
    "Predict Restaurant Trends", "Evaluate Regional Foot Traffic",
    "Compare Demographics", "Forecast Migration Rates",
    "Estimate Housing Demand", "Study Cuisine Popularity"
]

# Generate 72 unique research_query tuples
num_records = 72
research_query_data = []

# Ensure unique queryid and (dateofquery + timeofquery + userid)
used_query_ids = set()
used_date_time_user = set()

start_date = datetime(2020, 1, 1)  # Start of query period

while len(research_query_data) < num_records:
    # Generate a unique queryid
    queryid = random.randint(1, 1000000)
    if queryid in used_query_ids:
        continue

    # Select a random user (ensuring each user has 0-3 queries)
    userid = random.choice(user_ids)
    user_query_count = sum(1 for record in research_query_data if record['userid'] == userid)
    if user_query_count >= 3:
        continue

    # Generate a random date and time
    dateofquery = start_date + timedelta(days=random.randint(0, 1000))
    timeofquery = (datetime.min + timedelta(seconds=random.randint(0, 86399))).time()

    # Ensure (dateofquery + timeofquery + userid) is unique
    if (dateofquery, timeofquery, userid) in used_date_time_user:
        continue

    # Select a random query function
    queryfunction = random.choice(query_functions)

    # Add the tuple
    research_query_data.append({
        "queryid": queryid,
        "queryfunction": queryfunction,
        "dateofquery": dateofquery.date(),
        "timeofquery": timeofquery,
        "userid": userid
    })
    used_query_ids.add(queryid)
    used_date_time_user.add((dateofquery, timeofquery, userid))

# Convert to DataFrame
df_research_query = pd.DataFrame(research_query_data)

# Display the generated data
print(df_research_query.head())

# Export to CSV
csv_file_path = "research_query.csv"
df_research_query.to_csv(csv_file_path, index=False)
print(f"\nNew data exported to {csv_file_path}")
```

```python
import pandas as pd
import random
from google.colab import files

# Upload the country.csv and demographic_stat.csv files
uploaded = files.upload()

# Load the datasets
country_data = pd.read_csv('country.csv')
demographic_stat_data = pd.read_csv('demographic_stat.csv')

# Extract country names and demographic IDs
countries = country_data['country_name'].tolist()
demographic_ids = demographic_stat_data['id'].tolist()

# Generate 3000 unique Country_Ranks tuples
num_records = 3000
country_ranks_data = []

# Ensure uniqueness of (Country, DemographicID) and (DemographicID, Ranks)
used_country_demo = set()
used_demo_rank = {}

while len(country_ranks_data) < num_records:
    # Randomly select a demographic ID
    demographic_id = random.choice(demographic_ids)

    # Randomly select up to 10 countries for this demographic ID
    selected_countries = random.sample(countries, min(10, len(countries)))

    # Assign ranks from 1 to the number of selected countries
    for rank, country in enumerate(selected_countries, start=1):
        # Ensure (Country, DemographicID) is unique
        if (country, demographic_id) in used_country_demo:
            continue

        # Ensure (DemographicID, Ranks) is unique
        if demographic_id not in used_demo_rank:
            used_demo_rank[demographic_id] = set()

        if rank in used_demo_rank[demographic_id]:
            continue

        # Add the tuple
        country_ranks_data.append({
            "Country": country,
            "DemographicID": demographic_id,
            "Ranks": rank
        })

        # Track usage
        used_country_demo.add((country, demographic_id))
        used_demo_rank[demographic_id].add(rank)

        # Stop once 3000 records are generated
        if len(country_ranks_data) >= num_records:
            break
    if len(country_ranks_data) >= num_records:
        break

# Convert to DataFrame
df_country_ranks = pd.DataFrame(country_ranks_data)

# Display the generated data
print(df_country_ranks.head())

# Export to CSV
csv_file_path = "country_ranks.csv"
df_country_ranks.to_csv(csv_file_path, index=False)
print(f"\nNew data exported to {csv_file_path}")
```
